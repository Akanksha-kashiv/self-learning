```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
z=pd.read_csv("zomato.csv")
```

```python
z.head(8)
```

```python
z = z.applymap(lambda x: x.strip() if isinstance(x, str) else x)
z.head(20)
```

```python
z['RATING'].replace(to_replace='-',value=4.5,inplace=True)
z['RATING'].iloc[11330:11337]
```

```python
z['RATING'].replace(to_replace='NEW',value=5.0,inplace=True)
```

```python

# Plotting the number of restaurants for each 
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_context("paper", font_scale = 1.25, rc = {"font.size": 20,"axes.titlesize": 25,"axes.labelsize": 20})   
b = sns.catplot(data = z_orig, kind = 'count', x = 'RATING', order = z_orig['RATING'].value_counts().index)
plt.title('Number of restaurants for each rating')
b.set_xticklabels(rotation = 90)
plt.show()
```
